#2. Load dataset

#4. Sort dataset by ***nationality***

```python
# code
df_sorted = df.sort_values(by='nationality',ascending=True)

df_sorted
```

```python
# code
grouped_df = df.groupby(['nationality', 'gender']).size().reset_index(name='count')
# grouped_df = df.groupby(['nationality', 'gender']).groups
grouped_df
```

#6. Use scatter plot to represent the relationship between ***age*** and ***refletter.rating***
Rememer adding titles, xlabel, ylabel, ... to the plot

```python
# code
plt.scatter(df['age'], df['refletter.rating'], color='green')
plt.title('Relationship between Age and Reference Letter Rating')
plt.xlabel('Age')
plt.ylabel('Reference Letter Rating')
plt.grid(True)
plt.show()
```

#7. Use scatter plot to represent the relationship between ***age*** and ***sciences.grade***
Rememer adding titles, xlabel, ylabel, ... to the plot

```python
#code
plt.scatter(df['age'], df['sciences.grade'],color='green')
plt.title('Relationship between age and sciences.grade')
plt.xlabel('Age')
plt.ylabel('Grade')
plt.grid(True)
plt.show()
```

#8. Use histogram plot to represent the distribution of ***Age***

#9. Create a data frame that computes the average of ages in each countries?

Hint: Use groupby, select the age column and aggregate using mean.

```python
#code
avg_age_df = df.groupby('nationality')['age'].mean().reset_index(name='average_age')

avg_age_df
```

#10. Create a dataframe including average of ***english.grade, math.grade, sciences.grade, language.grade, portfolio.rating, coverletter.rating, refletter.rating*** of each country


```python
#code
avg_grades_df = df.groupby('nationality').agg({
    'english.grade': 'mean',
    'math.grade': 'mean',
    'sciences.grade': 'mean',
    'language.grade': 'mean',
    'portfolio.rating': 'mean',
    'coverletter.rating': 'mean',
    'refletter.rating': 'mean'
}).reset_index()

avg_grades_df
```
